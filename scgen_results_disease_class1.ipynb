```python
train = sc.read('/content/drive/My Drive/scGen/mouse_bullosa.h5ad')
```

```python
from matplotlib.pyplot import rcParams
import matplotlib.pyplot as plt
rcParams['figure.figsize'] = 5, 3
sc.pl.umap(train, color=['cell_type', 'disease'], legend_fontsize=8, show=False, frameon=False, ncols=1)
plt.tight_layout()
plt.show()
```

```python
train.obs["cell_type"].value_counts()
```

```python
train.obs["disease"].value_counts()
```

```python
import numpy as np

# Assuming your AnnData object is called adata
# Get the number of rows (observations)
num_obs = train.n_obs

# Define the number of rows to remove
num_to_remove = 351

# Generate random indices to remove
remove_indices = np.random.choice(num_obs, size=num_to_remove, replace=False)

# Keep all indices except the ones to remove
keep_indices = [i for i in range(num_obs) if i not in remove_indices]

# Subset the AnnData object using keep indices
train_new = train[keep_indices].copy()

# This creates a new AnnData object with the remaining data
print(f"New AnnData object with shape: {train_new.shape}")

train_removed = train[remove_indices].copy()

# This creates another AnnData object with just the removed rows
print(f"AnnData object containing the removed rows: {train_removed.shape}")
```

```python
from matplotlib.pyplot import rcParams
import matplotlib.pyplot as plt
rcParams['figure.figsize'] = 5, 3
sc.pl.umap(train_new, color=['cell_type', 'disease'], legend_fontsize=8, show=False, frameon=False, ncols=1)
plt.tight_layout()
plt.show()
```

```python
from matplotlib.pyplot import rcParams
import matplotlib.pyplot as plt
rcParams['figure.figsize'] = 5, 3
sc.pl.umap(train_removed, color=['cell_type', 'disease'], legend_fontsize=8, show=False, frameon=False, ncols=1)
plt.tight_layout()
plt.show()
```

```python
scgen.SCGEN.setup_anndata(train_new, batch_key="disease", labels_key="cell_type")
```

```python
model = scgen.SCGEN(train_new)
model.save("/content/drive/MyDrive/scGen/saved_models/Mouse_bullosa.pt", overwrite=True)
```

```python
model.train(
    max_epochs=300,
    batch_size=32,
    early_stopping=True,
    early_stopping_patience=300
)
```

```python
model.save("/content/drive/MyDrive/scGen/saved_models/Mouse_bullosa_100epochs.pt", overwrite=True)
```

```python
latent_X = model.get_latent_representation()
latent_adata = sc.AnnData(X=latent_X, obs=train_new.obs.copy())
latent_adata.write('/content/drive/MyDrive/scGen/Results/Mouse_bul/latentdata_100epochs.h5ad', compression='lzf')
```

```python
sc.settings.figdir = '/content/drive/MyDrive/scGen/Results/Mouse_bul'
```

```python
sc.pp.neighbors(latent_adata)
sc.tl.umap(latent_adata)
sc.pl.umap(latent_adata, color=['disease', 'cell_type'], wspace=0.4, frameon=False,
           save='latentspace_MB_100epochs.jpeg')
```
