```python
import logging
import scanpy as sc
import scgen
```

```python
!wget -O '/content/drive/My Drive/scGen/gut_crohns.h5ad' 'https://datasets.cellxgene.cziscience.com/a4e81c2f-2e1e-4a6c-810f-88aa7fce3125.h5ad'
```

```python
train = sc.read('/content/drive/My Drive/scGen/gut_crohns.h5ad')
```

```python
train.obs['disease'].value_counts()
```

```python
train.obs
```

```python
from matplotlib.pyplot import rcParams
import matplotlib.pyplot as plt
rcParams['figure.figsize'] = 5, 3
sc.pl.umap(train, color=['cell_type', 'disease'], legend_fontsize=8, show=False, frameon=False, ncols=1)
plt.tight_layout()
plt.show()
```

```python
plt.savefig('/content/drive/MyDrive/scGen/Results/Gut_crohns/with_fibroblast.jpeg')
```

```python
train_new = train[~((train.obs["cell_type"] == "fibroblast") &
                    (train.obs["disease"] == "Crohn disease"))].copy()
```

```python
from matplotlib.pyplot import rcParams
import matplotlib.pyplot as plt
rcParams['figure.figsize'] = 5, 3
sc.pl.umap(train_new, color=['cell_type', 'disease'], legend_fontsize=8, show=False, frameon=False, ncols=1)
plt.tight_layout()
plt.show()
plt.savefig('/content/drive/MyDrive/scGen/Results/Gut_crohns/without_fibroblast.jpeg')
```

```python
scgen.SCGEN.setup_anndata(train_new, batch_key="disease", labels_key="cell_type")
```

```python
model = scgen.SCGEN(train_new)
model.save("/content/drive/MyDrive/scGen/saved_models/Gut_crohns.pt", overwrite=True)
```

```python
model.train(
    max_epochs=100,
    batch_size=32,
    early_stopping=True,
    early_stopping_patience=100
)
```

```python
model.save("/content/drive/MyDrive/scGen/saved_models/Gut_crohns_100epochs.pt", overwrite=True)
```

```python
latent_X = model.get_latent_representation()
latent_adata = sc.AnnData(X=latent_X, obs=train_new.obs.copy())
latent_adata.write('/content/drive/MyDrive/scGen/Results/Gut_crohns/latentdata_100epochs.h5ad', compression='lzf')
```

```python
sc.settings.figdir = '/content/drive/MyDrive/scGen/Results/Gut_crohns'
```

```python
sc.pp.neighbors(latent_adata)
sc.tl.umap(latent_adata)
sc.pl.umap(latent_adata, color=['disease', 'cell_type'], wspace=0.4, frameon=False,
           save='latentspace_Gut_crohns_100epochs.jpeg')
```

```python
pred, delta = model.predict(
    ctrl_key='normal',
    stim_key='Crohn disease',
    celltype_to_predict='fibroblast'
)
pred.obs['disease'] = 'pred'
pred.write('/content/drive/MyDrive/scGen/Results/Gut_crohns_100epochs.h5ad', compression='lzf')
```

```python
ctrl_adata = train[((train.obs['cell_type'] == 'fibroblast') & (train.obs['disease'] == 'normal'))]
stim_adata = train[((train.obs['cell_type'] == 'fibroblast') & (train.obs['disease'] == 'Crohn disease'))]
```

```python
eval_adata = ctrl_adata.concatenate(stim_adata, pred)
```

```python
sc.tl.pca(eval_adata)
sc.pl.pca(eval_adata, color="disease", frameon=False,
           save='pred_stim_Gut_crohns_100epochs.jpeg')
```

```python
fib = train[train.obs["cell_type"] =="fibroblast"]
```

```python
sc.tl.rank_genes_groups(fib, groupby="disease", method="wilcoxon")
diff_genes = fib.uns["rank_genes_groups"]["names"]["Crohn disease"]
print(diff_genes)
```

```python
r2_value = model.reg_mean_plot(
    eval_adata,
    axis_keys={"x": "pred", "y": "Crohn disease"},
    gene_list=diff_genes[:10],
    labels={"x": "predicted", "y": "ground truth"},
    path_to_save="Crohn_disease_100epochs_reg_mean_100epochs.jpeg",
    show=True,
    legend=False
)
```

```python
sc.pl.violin(eval_adata, keys="ENSG00000211592", groupby="disease", save = 'Gut_crohns_100epochs_violin.jpeg')
```

```python
model.train(
    max_epochs=500,
    batch_size=32,
    early_stopping=True,
    early_stopping_patience=500
)
```
